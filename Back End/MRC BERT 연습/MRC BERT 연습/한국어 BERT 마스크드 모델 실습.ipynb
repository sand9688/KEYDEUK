```python
from transformers import TFBertForMaskedLM, AutoTokenizer
```

```python
model = TFBertForMaskedLM.from_pretrained('klue/bert-base', from_pt=True)# 파이토치에서 학습된 모딜이지만 텐서플로우에서 사용하겠다는 의미로 from_pt=True를 사용
tokenizer = AutoTokenizer.from_pretrained('klue/bert-base')
```

BERT 입력

```python
inputs = tokenizer('축구는 정말 재미있는 [MASK] 다.', return_tensors='tf')
```

```python
print(inputs['input_ids'])#정수 인코딩
```

```python
print(inputs['token_type_ids']) #세그먼트 인코딩
```

```python
print(inputs['attention_mask'])
```

```python
from transformers import FillMaskPipeline
```

```python
pip = FillMaskPipeline(model=model, tokenizer=tokenizer)
```

```python
pip('축구는 정말 재미있는 [MASK]다.')
```
