```python
import pandas as pd
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased") # Bert-base의 토크나이져
```

```python
result = tokenizer.tokenize('Hello my name is KK, How are you?')
print(result)
```

```python
print(tokenizer.vocab['안녕'])
```

```python
print(tokenizer.vocab['Helloworld'])
```

```python
print(tokenizer.vocab['##ding'])
```

```python
#bert의 단어 집합을 vocabulary.txt에 저장
with open('vocabulary.txt','w')as f:
  for token in tokenizer.vocab.keys():
    f.write(token+'\n')
```

```python
df = pd.read_fwf('vocabulary.txt', header=None)
df
```

```python
print('단어 집합의 크기 : ',len(df))
```
