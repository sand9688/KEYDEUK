마스크드 언어모델과 토크나이져

```python
from transformers import TFBertForMaskedLM, AutoTokenizer #TFBertForMaskedLM 과 AutoTokenizer 모듈을 불러온다
```

```python
model = TFBertForMaskedLM.from_pretrained('bert-large-uncased') # 모델생성
tokenizer = AutoTokenizer.from_pretrained('bert-large-uncased') # 토큰나이져 생성
```

Bert 입력

```python
#인코딩
inputs = tokenizer('Soccer is a really fun [MASK].', return_tensors='tf')
```

```python
#인코딩 결과 확인하기
print(inputs['input_ids'])
```

```python
#세그먼트 인코딩 
print(inputs['token_type_ids'])
```

```python
#어텐션 마스크 확인하기
print(inputs['attention_mask'])
```

[MASK] 토큰 예측하기 FillMaskPipeline 활용한

```python
from transformers import FillMaskPipeline
pip = FillMaskPipeline(model=model, tokenizer=tokenizer)
```

```python
pip('Soccer is a really fun [MASK].')
```
